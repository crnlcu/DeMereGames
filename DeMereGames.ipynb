```python
import numpy
import random
import matplotlib.pyplot as plt


def game1(total_iterations,seed=None):
    if seed is not None:
        random.seed(seed)
    die_faces = 6
    win_number = 6
    max_tosses = 4
    game_results = []
    wins_count = 0

    for i in range(1, total_iterations + 1):
        tosses = 0
        roll = 0
        while tosses < max_tosses and roll != win_number:
            roll = random.randint(1, die_faces)
            tosses += 1
            if roll == win_number:
                wins_count += 1
        game_results.append(wins_count / i)
    return game_results
def game2(total_iterations,seed=None):
    if seed is not None:
        random.seed(seed)
    die_faces = 6
    win_number = 6
    max_tosses = 24
    game_results = []
    wins_count = 0

    for i in range(1, total_iterations + 1):
        tosses = 0
        roll1 = 0
        roll2 = 0
        while tosses < max_tosses:
            roll1 = random.randint(1, die_faces)
            roll2 = random.randint(1, die_faces)
            tosses += 1
            
            if roll1 == win_number and roll2 == win_number: 
                wins_count += 1; 
                break
        
        game_results.append(wins_count / i)
        
        
    return game_results
```

```python

my_iterations=1000

seed=0


results_g1 = game1(my_iterations,seed)
results_g2 = game2(my_iterations,seed)

plt.plot(results_g1,color="black",lw=0.75,label="Game 1")
plt.plot(results_g2,color="navy",lw=0.5,label="Game 2")
plt.axhline(1 - (5/6)**4, color="gray", linestyle="--", lw=0.75, label="Theoretical Probability Game 1")
plt.axhline(1 - (35/36)**24, color="purple", linestyle="--",lw=0.5, label="Theoretical Probability Game 2" )
plt.xlabel("Number of Games")
plt.ylabel("Winning Probability")
plt.title("Convergence of Winning Probability - De Mere’s Games")
plt.xlim(0,my_iterations)
plt.legend()
plt.show()
```
